```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
from functions import load_and_clean_csv,check_nulls_and_duplicates,convert_columns_to_type
```

```python
url_1 = "datasets/heart_attack_prediction_dataset.csv"
df_1 = load_and_clean_csv(url_1)
df_1.head()
```

```python
url_2 = "datasets/pib_paises_en.csv"
df_2 = load_and_clean_csv(url_2)
df_2.head()
```

```python
tarject_columns_1 = ["age",
                     "sex",
                     "blood_pressure",
                     "heart_rate",
                     "obesity",
                     "exercise_hours_per_week",
                     "sedentary_hours_per_day",
                     "bmi",
                     "physical_activity_days_per_week",
                     "sleep_hours_per_day", "country"
                     ]
df_clean_1 = df_1[tarject_columns_1]
df_clean_1.head()
```

```python
tarject_columns_2 = ["country", 
                     "gdp",
                     ]
df_clean_2 = df_2[tarject_columns_2]
df_clean_2.head()
```

```python
merged_df = pd.merge(df_clean_1, df_clean_2, on= "country", how= "inner")
merged_df.sample(20)
```

```python
results = check_nulls_and_duplicates(merged_df)

print("Columns with null values")
print(results['nulls'])
print("\nHow many rows are duplicated: ", results['duplicates'])
```

```python
columns_to_convert = ["sedentary_hours_per_day", "exercise_hours_per_week", "bmi"]
desired_dtype = "int64"
merged_df = convert_columns_to_type(merged_df, columns_to_convert, desired_dtype)
merged_df.dtypes
```

```python
merged_df.to_csv('merged_data.csv', index=False)
```
